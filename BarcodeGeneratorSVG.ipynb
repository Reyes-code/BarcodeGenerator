```python
import barcode
import pandas as pd
from barcode.writer import ImageWriter, SVGWriter
```

```python
data =pd.read_excel("codigosbarrra.xlsx")
data
```

```python
def generar_codigo_de_barras(numero, formato):
    # Crea un objeto de código de barras EAN-128
    codigo_de_barras = barcode.get(formato, numero, writer=SVGWriter())

    # Guarda el código de barras en un archivo
    nombre_archivo = f"{numero}"
    codigo_de_barras.save(nombre_archivo)

    print(f"Se generó el código de barras en el archivo: {nombre_archivo}")
```

```python
for i in data["Código"]:
    numero_codigo = i
    formato_codigo = "code128"
    generar_codigo_de_barras(numero_codigo, formato_codigo)
```
